```python
# Carregar os dados do arquivo em um dicionário Python chamado data

with open('deliveries.json', mode='r', encoding='utf8') as file:
  data = json.load(file)
```

```python
# Retorna o número de elementos no objeto 'data'

len(data)
```

**3.2 Data Wrangling**

```python
# Criando um DataFrame

deliveries_df = pd.DataFrame(data)
deliveries_df.head()
```

```python
# normalizando a coluna origin com uma operação conhecida como flatten ou achatamento, que transforma cada chave do JSON em uma nova coluna:

hub_origin_df = pd.json_normalize(deliveries_df["origin"])
hub_origin_df.head()
```

```python
# Realizando a mesclagem (merge) dos DataFrames deliveries_df e hub_origin_df
# Usando a junção do tipo 'inner' (apenas registros que possuem correspondência em ambos os DataFrames serão mantidos)
# A junção é feita com base nos índices de ambos os DataFrames

deliveries_df = pd.merge(left=deliveries_df, right=hub_origin_df, how='inner', left_index=True, right_index=True)
deliveries_df.head()
```

```python
# Removendo a coluna "origin" do DataFrame, pois não é mais necessária
deliveries_df = deliveries_df.drop("origin", axis=1)

# Reorganizando as colunas do DataFrame na seguinte ordem:
# "name" (nome), "region" (região), "lng" (longitude), "lat" (latitude),
# "vehicle_capacity" (capacidade do veículo) e "deliveries" (quantidade de entregas)
deliveries_df = deliveries_df[["name", "region", "lng", "lat", "vehicle_capacity", "deliveries"]]

# Exibindo as primeiras 5 linhas do DataFrame para verificar as alterações
deliveries_df.head()
```

```python
# Renomeando as colunas "lng" e "lat" para "hub_lng" e "hub_lat", respectivamente, para indicar que essas coordenadas pertencem ao hub de origem
deliveries_df.rename(columns={"lng": "hub_lng", "lat": "hub_lat"}, inplace=True)
deliveries_df.head()
```

```python
# Criando um novo DataFrame contendo apenas a coluna "deliveries"
# A função explode() transforma os elementos de listas dentro da coluna "deliveries" em múltiplas linhas, mantendo a relação com o índice original
deliveries_exploded_df = deliveries_df[["deliveries"]].explode("deliveries")
deliveries_exploded_df.head()
```

```python
# Criando um novo DataFrame a partir da coluna "deliveries" do deliveries_exploded_df
# Cada campo aninhado dentro de "deliveries" será extraído e transformado em colunas separadas

deliveries_normalized_df = pd.concat([ pd.DataFrame(deliveries_exploded_df ["deliveries"].apply( lambda record: record["size"]) ).rename(columns={"deliveries": "delivery_size"}),
                                      pd.DataFrame(deliveries_exploded_df ["deliveries"].apply( lambda record: record["point"]["lng"]) ).rename(columns={"deliveries":  "delivery_lng"}),
                                       pd.DataFrame(deliveries_exploded_df ["deliveries"].apply( lambda record: record["point"]["lat"]) ).rename(columns={"deliveries":  "delivery_lat"}), ],
                                     axis= 1) # Concatenando as colunas lado a lado
deliveries_normalized_df .head()
```

```python
# Removendo a coluna "deliveries" do DataFrame, já que ela não é mais necessária
deliveries_df = deliveries_df.drop("deliveries", axis=1)

# Realizando um merge com o DataFrame 'deliveries_normalized_df', usando as colunas de índice de ambos os DataFrames
# A junção é feita pela chave de índice, mantendo todas as linhas do 'deliveries_normalized_df' (right join)
deliveries_df = pd.merge(left=deliveries_df, right=deliveries_normalized_df, how='right', left_index=True, right_index=True)

# Reseta o índice do DataFrame, criando um novo índice numérico contínuo e descartando o índice anterior
deliveries_df.reset_index(inplace=True, drop=True)
deliveries_df.head()
```

# **4. Manipulação**

```python
# Geocodificação reversa do hub

hub_df = deliveries_df[["region", "hub_lng", "hub_lat"]]
hub_df = hub_df.drop_duplicates().sort_values(by="region").reset_index( drop=True)

hub_df.head()
```

```python
#Cconvertendo as coordenadas de latitude e longitude em informações de localização

geolocator = Nominatim(user_agent="ebac_geocoder")
location = geolocator.reverse("-15.657013854445248, -47.802664728268745" )

print(json.dumps(location.raw, indent=2, ensure_ascii=False))
```

```python
# Aplicar a geocodificação nas coordenadas das três regiões e extrair informações de cidade e bairro.

geocoder = RateLimiter(geolocator.reverse, min_delay_seconds=1)
hub_df.head()
```

```python
# Cria uma nova coluna 'coordinates' no DataFrame 'hub_df',
#combinando as colunas 'hub_lat' e 'hub_lng' em uma string formatada (latitude, longitude)
hub_df["coordinates"] = hub_df["hub_lat"].astype(str) + ", "
hub_df["coordinates"] = hub_df["coordinates"] + hub_df["hub_lng"].astype(str)
hub_df["geodata"] = hub_df["coordinates"].apply(geocoder)

hub_df.head()
```

```python
hub_geodata_df = pd.json_normalize(hub_df["geodata"].apply( lambda data: data.raw ))

hub_geodata_df.head()
```

```python
# Filtrando o DataFrame

hub_geodata_df = hub_geodata_df[ ["address.town", "address.suburb", "address.city"] ]
```

```python
# Renomeando as colunas para um entendimento mais claro

hub_geodata_df.rename(columns = {"address.town":"hub_town",
                                 "address.suburb":"hub_suburb",
                                 "address.city":"hub_city"},
                      inplace=True)
# Substituindo valores nulos

hub_geodata_df["hub_city"] = np.where(hub_geodata_df["hub_city"].notna(), hub_geodata_df["hub_city"], hub_geodata_df["hub_town"])
hub_geodata_df["hub_suburb"] = np.where( hub_geodata_df["hub_suburb"].notna(), hub_geodata_df["hub_suburb"], hub_geodata_df["hub_city"])

# Removendo a coluna 'hub_town' após seu uso, já que a informação foi substituida por 'hub_city'

hub_geodata_df = hub_geodata_df.drop("hub_town", axis=1)

hub_geodata_df.head()
```

```python
# Mesclando o DataFrame 'hub_df' com 'hub_geodata_df' usando os índices como chave

hub_df = pd.merge( left=hub_df, right=hub_geodata_df, left_index=True, right_index=True )
hub_df = hub_df[["region", "hub_suburb", "hub_city"]]
hub_df.head()
```

```python
# Realiza um merge entre 'deliveries_df' e 'hub_df', utilizando a coluna 'region' como chave de junção.
# O tipo de junção é 'inner', o que significa que apenas as linhas com correspondência em ambas as tabelas serão mantidas.
deliveries_df = pd.merge(left=deliveries_df, right=hub_df, how="inner", on="region")

# Seleciona um subconjunto das colunas para manter no DataFrame final,
# focando nas informações relevantes para a análise
deliveries_df = deliveries_df[[
    "name",               # Nome do entregador ou da entrega
    "region",             # Região associada à entrega
    "hub_lng",            # Longitude do hub de distribuição
    "hub_lat",            # Latitude do hub de distribuição
    "hub_city",           # Cidade associada ao hub
    "hub_suburb",         # Subúrbio ou bairro do hub
    "vehicle_capacity",   # Capacidade do veículo
    "delivery_size",      # Tamanho da entrega
    "delivery_lng",       # Longitude da entrega
    "delivery_lat"        # Latitude da entrega
]]

deliveries_df.head()
```

# **5. Visualização**

**5.1 Mapa do Distrito Federal**

```python
# Download dos dados do mapa do Distrito Federal do site do IBGE no link https://www.ibge.gov.br/geociencias/cartas-e-mapas/bases-cartograficas-continuas para criar o DataFrame mapa.
!wget -q https://geoftp.ibge.gov.br/cartas_e_mapas/bases_cartograficas_continuas/bc100/go_df/versao2016/shapefile/bc100_go_df_shp.zip -O distrito-federal.zip
!unzip -q distrito-federal.zip -d ./maps
!cp ./maps/LIM_Unidade_Federacao_A.shp ./distrito-federal.shp
!cp ./maps/LIM_Unidade_Federacao_A.shx ./distrito-federal.shx
```

```python
# Carregando o arquivo shapefile contendo os dados geoespaciais do Distrito Federal
mapa = geopandas.read_file("distrito-federal.shp")

# Filtrando o DataFrame para selecionar apenas a primeira linha
mapa = mapa.loc[[0]]

# Exibindo o DataFrame resultante
mapa.head()
```

**5.2 Mapa Hubs**

```python
# Selecionando as colunas 'region', 'hub_lng' (longitude) e 'hub_lat' (latitude) do DataFrame 'deliveries_df',
# removendo as linhas duplicadas e reinicia o índice
hub_df = deliveries_df[["region", "hub_lng", "hub_lat"]].drop_duplicates().reset_index(drop=True)

# Criando um GeoDataFrame a partir do DataFrame 'hub_df', convertendo as colunas de longitude e latitude em um objeto geométrico (pontos)
geo_hub_df = geopandas.GeoDataFrame(hub_df, geometry=geopandas.points_from_xy(hub_df["hub_lng"], hub_df["hub_lat"]))

geo_hub_df.head()
```

**5.3 Mapa das Entregas**

```python
# Criando um GeoDataFrame a partir do DataFrame 'deliveries_df', convertendo as colunas de longitude e latitude da entrega em um objeto geométrico (pontos)

geo_deliveries_df = geopandas.GeoDataFrame( deliveries_df, geometry=geopandas.points_from_xy( deliveries_df["delivery_lng"], deliveries_df["delivery_lat"] ) )
geo_deliveries_df.head()
```

```python
import matplotlib.pyplot as plt

# Cria o plot vazio
fig, ax = plt.subplots(figsize = (20, 20))

# Plot mapa do Distrito Federal
mapa.plot(ax=ax, alpha=0.4, color="lightgrey")

# Plot das entregas
geo_deliveries_df.query("region == 'df-0'").plot( ax=ax, markersize=10, color="red", label="df-0" )
geo_deliveries_df.query("region == 'df-1'").plot( ax=ax, markersize=10, color="blue", label="df-1" )
geo_deliveries_df.query("region == 'df-2'").plot( ax=ax, markersize=10, color="seagreen", label="df-2" )

# Plot dos Hubs
geo_hub_df.plot( ax=ax, markersize=30, marker="x", color="black", label="hub" )

# Plot da legenda
plt.title( "Entregas no Distrito Federal por Região" , fontdict={"fontsize": 16} )
lgnd = plt.legend(loc='upper right')
for handle in lgnd.legend_handles:
  handle.set_sizes([50])
```

De acordo com o mapa, as entregas estão devidamente distribuídas entre os hubs correspondentes. No entanto, os hubs das regiões 0 e 2 realizam entregas em áreas distantes do centro e entre si, o que pode aumentar o tempo e os custos logísticos. Uma reavaliação da alocação de veículos ou da localização dos hubs pode contribuir para otimizar a eficiência das entregas.
